```python
import re

# 1. Read the original file
with open('expanded_kb.ttl', 'r', encoding='utf-8') as file:
    content = file.read()

# 2. Use Regex to find and replace <wd:Q...> with wd:Q...
# This looks for '<', followed by 'wd:', followed by any characters that aren't '>', followed by '>'
cleaned_content = re.sub(r'<wd:([^>]+)>', r'wd:\1', content)

# 3. Save the cleaned content to a new file (or overwrite the old one)
with open('expanded_kb_cleaned.ttl', 'w', encoding='utf-8') as file:
    file.write(cleaned_content)

print("File cleaned successfully!")
```
